# Predicting heart disease through Machine Learning

Models Used : Binary Classification (Linear Regression) , Random Forest , Random Forest GridSearch, Random Forest RandomisedSearch, Support Vector Machine

```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
from sklearn.ensemble import RandomForestClassifier 
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import export_graphviz
from sklearn import tree, metrics

plt.show()
sb.set() # set the default Seaborn style for graphics

dataframe = pd.read_excel('HeartDiseaseCleaned.xlsx', index_col=[0])

#Change all the >= 1 for number to 1 (Explained in EDA)
dataframe['number'].values[dataframe['number'].values > 0] = 1

#Encode categorical variables into numerics
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

sex_map = {'Male': 0, 'Female': 1}
cp_map = {'typical angina': 0, 'asymptomatic': 1 , 'non-anginal' : 2 , 'atypical angina' : 3}
exan_map = {False: 0, True: 1}
diabetic_map = {False: 0, True: 1}
restecg_map = {'lv hypertrophy': 0, 'normal': 1 , 'st-t abnormality' : 2}

dataframe['sex'] = dataframe['sex'].map(sex_map)
dataframe['chest_pain'] = dataframe['chest_pain'].map(cp_map)
dataframe['exercise_angina'] = dataframe['exercise_angina'].map(exan_map)
dataframe['diabetic'] = dataframe['diabetic'].map(diabetic_map)
dataframe['resting_ecg'] = dataframe['resting_ecg'].map(restecg_map)

dataframe
```

Ensuring there is no NaN before proceeding

```python
dataframe.isnull().sum()
```

```python
dataframe.describe()
```

## Splitting the dataset - used for all prediction models

```python
#split dataset in features and target variable
feature_cols = ['age', 'sex', 'chest_pain', 'rest_blood_pressure','cholesterol_level','diabetic','resting_ecg', 'max_heart_rate', 'exercise_angina']
X = dataframe[feature_cols] # Features
y = dataframe.number # Target variable

# split X and y into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle = False)
```

## **Data Analysis through machine learning : Binary Classification**
#### **Logistic regression** is used in **Binary Classification** to predict a **true or false outcome (1 or 0)**
#### Logistic regression is useful for binary related predictions, in this case we are predicing whether a person has heart disease or not : 1 or 0.

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(random_state=16, max_iter = 600)

logreg.fit(X_train, y_train)

coefficients = logreg.coef_[0]

y_pred = logreg.predict(X_test)

from sklearn import metrics

cnf_matrix = metrics.confusion_matrix(y_test, y_pred)

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

class_names=["No Heart Disease" , "Has heart Disease"] 
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)

sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="RdBu" ,fmt='g')
ax.xaxis.set_label_position("top")

plt.tight_layout()
plt.title('Confusion matrix (Binary Prediction)', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')

from sklearn.metrics import classification_report
target_names = ['Without heart disease', 'with heart disease']
print(classification_report(y_test, y_pred, target_names=target_names))

TP = cnf_matrix[1, 1]
FP = cnf_matrix[0, 1]
FN = cnf_matrix[1, 0]
TN = cnf_matrix[0, 0]

print ("Confusion Matrix Binary Classification |TP" ,  TP , "|FP" ,  FP ," |FN", FN , "|TN", TN )

print("Rates : |TPR" , round((TP / (TP + FN)),2) ,  "|TNR" , round(TN / ( TN + FP),2) , 
              "|FPR" , round(FP / ( FP + TN),2) , "|FNR" , round(FN / ( FN + TP),2))

plt.show()
```

From Logistic regression, we are able to predict the presence of heart disease **accurately 73%** of the time, where the **true positive rate is 0.76** , the **false positive rate is 0.45**, the **false negative rate is 0.24**

```python
odds_ratios = np.exp(coefficients)

feature_odds_ratios = {feature: odds_ratio for feature, odds_ratio in zip(X_train.columns, odds_ratios)}

sorted_feature_odds_ratios = dict(sorted(feature_odds_ratios.items(), key=lambda item: item[1], reverse=True))

print("\nStrength of variable association with the presence of heart disease in descending order")

for feature, odds_ratio in sorted_feature_odds_ratios.items():
    print(f"{feature}: {odds_ratio:.4f}")

features = list(sorted_feature_odds_ratios.keys())[::-1]
odds_ratios = list(sorted_feature_odds_ratios.values())[::-1]

plt.figure(figsize=(10, 6))
plt.barh(features, odds_ratios, color='skyblue')
plt.xlabel('Odds Ratio')
plt.ylabel('Feature')
plt.title('Odds Ratios of Predictors for Heart Disease')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

# Evaluating Binary Classification With Receiver Operating Characteristic(ROC) curve

```python
y_pred_proba = logreg.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)

#Plot y=x for visualisation of "midpoint - 0.5 accuracy"
x = np.linspace(0, 1)
y = x

#Plot Y=X to visualise what is 0.5 accuracy
plt.plot(x, y, linewidth=1.0 ,color="red", label="Y = X (Above = good, Below = bad)", linestyle=':')

#Plot FPR against TPR
plt.plot(fpr,tpr, color ="blue", label="Area Under Curve (AUC) = " + str(round(auc , 6)))
plt.legend(loc="upper left")

#AUC Classifier of 0.67136 - 67% accuracyÂ 
plt.show()
```

#### The blue line plots **True Postive Rate (TPR)** against **False Positive Rate (FPR)** - Shows the relation between TPR and FPR where the Area Under the Curve (AUC) is **0.70**
#### The red dotted line represents the seperation of a good and prediction ( AUC = 0.5 ). Above the line represents better than random prediction, while below the line represents worse than random prediction.
#### In our case, the AUC is above the line, meaning we are able to predict heart disease better than guessing at random


# Predicting heart disease with Random Forest
A better model as compared to a single decision tree. It also solves the problem of overfitting. It can also effective handle categorical and numerical variable.

```python
y = dataframe['number'] 
x = dataframe.drop(['number'], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, shuffle = False)
rfModel = RandomForestClassifier()
rfModel.fit(X_train, y_train)
y_pred = rfModel.predict(X_test)
feature_names = [f"{x.columns[i]}" for i in range(x.shape[1])]
print("Random Forest Default Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.datasets import make_classification
importances = rfModel.feature_importances_
forest_importances = pd.Series(importances, index=feature_names)
std = np.std([tree.feature_importances_ for tree in rfModel.estimators_], axis=0)

fig, ax = plt.subplots()
forest_importances.sort_values(ascending=False).plot.bar(yerr=std, ax=ax)
ax.set_title("Feature importances using MDI")
ax.set_ylabel("Mean decrease in impurity")
fig.tight_layout()
```

Base on the variable importance, chest pain, age, max heart rate, cholesterol level, blood pressure and exercise angina plays a larger role in predicting heart disease. While resting ecg, sex and diabetic plays a less important role.

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize = (5,5))
sb.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')

TP = cm[1, 1]
FP = cm[0, 1]
FN = cm[1, 0]
TN = cm[0, 0]
TPR = round((TP / (TP + FN)), 2)
FPR = round((FP / ( FP + TN)), 2)
print ("Confusion Matrix (Random Forest) |TP" ,  TP , "|FP" ,  FP ," |FN", FN , "|TN", TN)
print("Rates : |TPR" , round((TP / (TP + FN)),2) ,  "|TNR" , round(TN / ( TN + FP),2) , 
              "|FPR" , round(FP / ( FP + TN),2) , "|FNR" , round(FN / ( FN + TP),2))
```

### ***Random Forest Hyperparameter tuning, RandomizedSearchCV***
We try to improve our default parameters for random forest through hyperparameter tuning, which tweak model performance for better results for our dataset.
Randomisedsamples hyperparameters randomly to find the best-performing model in machine learning

```python
from sklearn.model_selection import GridSearchCV

# num of decision tree in random forest (default 100)
n_estimators = [20,40,60,80,100,120]

# num of features at every split (default sqrt)
max_features = ['auto', 'sqrt']

# max level of tree (default None)
max_depth = [2,4, 8, None]

# number of samples (default None)
max_samples = [0.5, 0.75, 1.0, None]

# min num of sample to split a node (default 2)
min_samples_split = [2, 5]

# min num of sample at leaf node (default 1)
min_samples_leaf = [1, 2]

# if bootstrap sample used (default True)
bootstrap = [True, False]

#store above together as a dict
param_grid = {'n_estimators': n_estimators,
             'max_features': max_features,
             'max_depth': max_depth,
             'max_samples': max_samples,
             'min_samples_split': min_samples_split,
             'min_samples_leaf': min_samples_leaf,
             'bootstrap': bootstrap}
```

```python
from sklearn.model_selection import RandomizedSearchCV
rs = RandomizedSearchCV(estimator = rfModel, param_distributions=param_grid, cv=10, verbose=2, n_jobs=4)
```

```python
rs.fit(X_train, y_train)
```

```python
print(rs.best_params_)
rfRs = rs.best_estimator_
rfRs.fit(X_train, y_train)
y_pred = rfRs.predict(X_test)
print("Random Forest RandomSearch Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize = (5,5))
sb.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')

TP = cm[1, 1]
FP = cm[0, 1]
FN = cm[1, 0]
TN = cm[0, 0]
TPR = round((TP / (TP + FN)), 2)
FPR = round((FP / ( FP + TN)), 2)
print ("Confusion Matrix (RF RandomSearch)|TP" ,  TP , "|FP" ,  FP ," |FN", FN , "|TN", TN)
print("Rates : |TPR" , round((TP / (TP + FN)),2) ,  "|TNR" , round(TN / ( TN + FP),2) , 
              "|FPR" , round(FP / ( FP + TN),2) , "|FNR" , round(FN / ( FN + TP),2))
```

```python
importances = rfRs.feature_importances_
forest_importances = pd.Series(importances, index=feature_names)
std = np.std([tree.feature_importances_ for tree in rfRs.estimators_], axis=0)

fig, ax = plt.subplots()
forest_importances.sort_values(ascending=False).plot.bar(yerr=std, ax=ax)
ax.set_title("Feature importances using MDI")
ax.set_ylabel("Mean decrease in impurity")
fig.tight_layout()
```

Comparing variable importance with default random forest, they have similar variable used to predict heart disease however, we can see that random search yield better overall prediction accuracy.

### ***Random Forest Hyperparameter tuning, GridSearchCV***

Grid search is used for systematically exploring different combinations of hyperparameters which helps optimize the Random Forest Model. <br>
Compring with randomised search, grid serach systematically explores different hyperparameter combinations to find best parameter for our Random Forest Model

```python
rfModel = RandomForestClassifier()
gs = GridSearchCV(estimator = rfModel, param_grid = param_grid, cv=10, verbose=2, n_jobs=4)
```

```python
gs.fit(X_train, y_train)
```

```python
rfGs = gs.best_params_
print(rfGs)
rfGs = gs.best_estimator_
rfGs.fit(X_train, y_train)
y_pred = rfGs.predict(X_test)
print("Random Forest GridSearch Accuracy:", metrics.accuracy_score(y_test, y_pred))
```
